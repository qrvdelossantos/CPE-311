```python
import random
from collections import Counter
random.seed(0)
salaries = [round(random.random()*1000000, -3) for _ in range(100)]
```

```python
#Mean
mean = sum(salaries) / len(salaries)
```

```python
#Mode
counts = Counter(salaries)
mode_val = max(counts, key=counts.get)
```

```python
#Median
sorted_salaries = sorted(salaries)
n = len(sorted_salaries)
if n % 2 == 0:
    median_val = (sorted_salaries[n//2 - 1] + sorted_salaries[n//2]) / 2
else:
    median_val = sorted_salaries[n//2]
```

```python
mean_val = sum(salaries) / n
variance_val = sum((x - mean_val)**2 for x in salaries) / (n - 1)
```

```python
#
std_dev_val = variance_val
std_dev_val = variance_val**0.5
```

```python
print(salaries)
print(mean_val)
print(mode_val)
print(median_val)
print(std_dev_val)
```

```python
average = sum(salaries)/len(salaries)
len_minus_one = len(salaries) - 1

x_xmean_squared = []
for salary in salaries:
    temp = (salary - average) ** 2
    x_xmean_squared.append(temp)

summation = sum(x_xmean_squared)
variance = summation / len_minus_one
print(variance)
```

```python
import pandas as pd
import numpy as np
df = pd.read_csv("diabetes.csv")
```

```python

print("Column Names:", df.columns.tolist())

print("Data Types:\n", df.dtypes)

print("Total Records:", len(df))

print("First 20 Records:\n", df.head(20))

print("Last 20 Records:\n", df.tail(20))

df.rename(columns={"Outcome": "Diagnosis"}, inplace=True)
```

```python

df["Classification"] = df["Diagnosis"].apply(lambda x: "Diabetes" if x == 1 else "No Diabetes")

withDiabetes = df[df["Diagnosis"] == 1]

noDiabetes = df[df["Diagnosis"] == 0]

Pedia = df[df["Age"] <= 19]

Adult = df[df["Age"] > 19]

avg_age = np.mean(df["Age"])
avg_glucose = np.mean(df["Glucose"])
print("Average Age:", avg_age)
print("Average Glucose:", avg_glucose)
```

```python
median_age = np.median(df["Age"])
median_glucose = np.median(df["Glucose"])
print("Median Age:", median_age)
print("Median Glucose:", median_glucose)

```

```python
middle_age = np.percentile(df["Age"], 50)
middle_glucose = np.percentile(df["Glucose"], 50)
print("Middle Age:", middle_age)
print("Middle Glucose:", middle_glucose)

```

```python
std_skin = np.std(df["SkinThickness"])
print("Standard Deviation of SkinThickness:", std_skin)
```
