```python
n = 4
src = 'A'
aux = 'C'
des = 'B'

def move_disks(n, src, aux, des):
    if n == 1:
        print(f'Move disk 1 from rod {src} to rod {des}')
        return

    move_disks(n-1, src, des, aux)
    print(f'Move disk {n} from rod {src} to rod {des}')
    move_disks(n-1, aux, src, des)

#Stage 1:
print("Stage 1:")
print("source: A")
print("auxiliary: C")
print("destination: B")
print("Move:")
move_disks(n, src, aux, des)



#Stage 2:
print("Stage 2:")
print("source: C")
print("auxiliary: A")
print("destination: B")
print("Move:")

move_disks(n, src, aux, des)


#Stage 3:
print("Stage 3:")
print("source: B")
print("auxiliary: A")
print("destination: C")
move_disks(n, src, aux, des)



```

Observation:<br>
1. Recursive Nature<br>
2. Exponential Growth<br>
3. Optimal Solution<br>



Explanation:<br>
1. Move n-1 disks from the source rod to auxiliary rod.<br>
2. Move the largest disk directly to the destination rod
3. Move the n-1 disks from the auxiliary rod to the destination rod

Conclusion:<br>
The tower of hanoi algorithm demonstrates the power of recursion in solving complex problems by breaking them into simpler ones.
