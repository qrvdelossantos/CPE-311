```python
# Step 2: Import libraries
import pandas as pd
import numpy as np

# Step 3: Load your CSVs from the shared folder
# Adjust the path to match your folder and filenames
earthquakes = pd.read_csv('/content/drive/MyDrive/earthquakes.csv')
faang = pd.read_csv('/content/drive/MyDrive/faang.csv', parse_dates=['date'])

# Quick check
print("Earthquakes sample:\n", earthquakes.head())
```

```python
# 1. Japan quakes
japan_quakes = earthquakes[
    (earthquakes['place'].str.contains('Japan', case=False, na=False)) &
    (earthquakes['magType'] == 'mb') &
    (earthquakes['mag'] >= 4.9)
]

# 2. Binning magnitudes
ml_quakes = earthquakes[earthquakes['magType'] == 'ml']
bins = range(0, int(ml_quakes['mag'].max()) + 2)
ml_quakes['mag_bin'] = pd.cut(ml_quakes['mag'], bins=bins)
bin_counts = ml_quakes['mag_bin'].value_counts().sort_index()

# 3. Monthly resample
faang['date'] = pd.to_datetime(faang['date'])
faang.set_index('date', inplace=True)
monthly = faang.groupby('ticker').resample('M').agg({
    'open': 'mean',
    'high': 'max',
    'low': 'min',
    'close': 'mean',
    'volume': 'sum'
})

# 4. Crosstab
crosstab = pd.crosstab(
    earthquakes['tsunami'],
    earthquakes['magType'],
    values=earthquakes['mag'],
    aggfunc='max'
)

# 5. Rolling 60-day aggregations
rolling = faang.groupby('ticker').rolling(60).agg({
    'open': 'mean',
    'high': 'max',
    'low': 'min',
    'close': 'mean',
    'volume': 'sum'
})

# 6. Pivot table
pivot = faang.pivot_table(
    index='ticker',
    values=['open', 'high', 'low', 'close', 'volume'],
    aggfunc='mean'
)

# 7. Z-scores for NFLX
nflx = faang[faang['ticker'] == 'NFLX']
z_scores = nflx.apply(lambda x: (x - x.mean()) / x.std()
                      if np.issubdtype(x.dtype, np.number) else x)

# 8. Merge events
events = pd.DataFrame({
    'ticker': ['FB'] * 3,
    'date': pd.to_datetime(['2018-07-25', '2018-03-19', '2018-03-20']),
    'event': [
        'Disappointing user growth announced after close.',
        'Cambridge Analytica story',
        'FTC investigation'
    ]
}).set_index(['date', 'ticker'])

merged = faang.reset_index().set_index(['date', 'ticker']).merge(
    events, how='outer', left_index=True, right_index=True
)

# 9. Indexed FAANG data
indexed = faang.groupby('ticker').transform(lambda x: x / x.iloc[0])

# Print results
print("Japan Quakes:\n", japan_quakes.head())
print("\nMagnitude Bins:\n", bin_counts)
print("\nMonthly Aggregations:\n", monthly.head())
print("\nCrosstab:\n", crosstab)
print("\nRolling Aggregations:\n", rolling.head())
print("\nPivot Table:\n", pivot)
print("\nNetflix Z-scores:\n", z_scores.head())
print("\nMerged Events:\n", merged.head())
print("\nIndexed FAANG Data:\n", indexed.head())
```
